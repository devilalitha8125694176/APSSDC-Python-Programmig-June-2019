```python
import pandas as pd
df=pd.read_csv('DataFiles/income.csv')
years=df.columns[2:]
incomeArkansas=df.values[3,2:]
plt.title('Income of Akansas')
plt.xlabel('Year')
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```

```python
# Income of All States from 2015 to 2013
#values-->rows
incomeallstates=[]
for i in range(2,len(df.columns)):
    print(i)
```

```python
# Income of All States from 2015 to 2013
#values-->rows
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates)
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('All States Income')
plt.grid()
plt.show()
```

```python
# Income of All States from 2015 to 2013
#values-->rows
incomeallstates=[]
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
plt.plot(years,incomeallstates,'yo')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('All States Income')
plt.grid()
plt.show()
```

```python
# State with highest average income in the last three years
incomelast3states=[]
for i in df.columns[-3:]:
    incomelast3states.append(df[i].mean())
plt.plot(years,incomelast3states,'yo')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('All States Income')
plt.grid()
plt.show()
```
